```python
import duckdb
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
con = duckdb.connect(":default:")
```

```python
con.sql("install postgres;")
con.sql("load postgres;")
```

```python
con.sql(f"""
CREATE SECRET postgres_secret (
    TYPE postgres,
    HOST {os.getenv("host")},
    PORT {os.getenv("port")},
    DATABASE {os.getenv("database")},
    USER {os.getenv("username")},
    PASSWORD {os.getenv("password")}
);
""")
```

```python
con.sql("ATTACH '' AS dvd_rental (TYPE postgres, SECRET postgres_secret, SCHEMA 'public');")
```

```python
con.sql("select * from dvd_rental.public.actor limit 15;")
```

```python
con.sql(
"""
select
  f.actor_id,
  a.first_name,
  a.last_name,
  count(f.film_id) as films
from
  dvd_rental.public.film_actor as f
left join 
  dvd_rental.public.actor as a
on f.actor_id = a.actor_id
group by
  f.actor_id,
  a.first_name,
  a.last_name,
order by
 films desc
"""
)
```

```python
con.sql(
"""
select
  f.film_id,
  f.title,
  count(a.actor_id) as actors
from
  dvd_rental.public.film as f
left join
  dvd_rental.public.film_actor as a
on
 f.film_id = a.film_id
group by
  f.film_id,
  f.title,
order by actors desc;
"""
)
```

```python
con.sql(
"""
with 
  actors_films as (
    select
        f.film_id,
        f.title,
        count(a.actor_id) as actors
    from
        dvd_rental.public.film as f
    left join
        dvd_rental.public.film_actor as a
    on
        f.film_id = a.film_id
    group by
        f.film_id,
        f.title
    order by 
        actors desc
)
select
  a.*,
  c.name
from actors_films as a
left join dvd_rental.public.film_category as fc
  on a.film_id = fc.film_id
left join dvd_rental.public.category as c
  on fc.category_id = c.category_id
"""
)
```

```python
con.sql(
"""
select
  c.name,
  count(fc.film_id) as films
from
  dvd_rental.public.category as c
left join
  dvd_rental.public.film_category as fc
on
 c.category_id = fc.category_id
group by
  c.category_id,
  c.name
order by
  films desc;
"""
)
```

```python
con.sql(
"""
select
  f.title,
  c.name,
  count(r.rental_id) as rentals_per_movie
from
  dvd_rental.public.film as f
join
  dvd_rental.public.inventory as i
on
  f.film_id = i.film_id 
join
  dvd_rental.public.rental as r
on
  i.inventory_id = r.inventory_id 
left join
  dvd_rental.public.film_category as fc
on
  f.film_id = fc.film_id 
left join
  dvd_rental.public.category as c
on
  fc.category_id = c.category_id 
group by
  f.film_id,
  f.title,
  c.name
order by
  rentals_per_movie desc
"""
)
```

```python
con.sql(
"""
select
  c.name,
  count(r.rental_id) as rentals_per_category
from dvd_rental.public.category as c
join
  dvd_rental.public.film_category as fc
on
  c.category_id = fc.category_id
join
  dvd_rental.public.inventory as i
on
  fc.film_id = i.film_id
join
  dvd_rental.public.rental as r
on
  i.inventory_id = r.inventory_id
group by
  c.name
order by 
 rentals_per_category desc;
"""
)
```

```python
con.sql("""
select
  c.name,
  avg(f.rental_rate)::decimal(4,2) as average_category_rental_rate,
from
  dvd_rental.public.category as c
join
  dvd_rental.public.film_category as fc
on
  c.category_id = fc.category_id
join
  dvd_rental.public.film as f
on
  fc.film_id = f.film_id
group by
  c.name
order by
  average_category_rental_rate desc;
""")
```

```python
con.sql("""
select
  c.name,
  avg(f.length)::integer as average_category_length,
from
  dvd_rental.public.category as c
join
  dvd_rental.public.film_category as fc
on
  c.category_id = fc.category_id
join
  dvd_rental.public.film as f
on
  fc.film_id = f.film_id
group by
  c.name
order by
  average_category_length desc;
""")
```
